##Support Vector Classifier

##Importing Necessary libraries
###digit recognization in sklearn datasets

```
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt
import matplotlib.cm as cm 
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
```

##Loading Datasets


```
digits = datasets.load_digits()
data = digits.data
target = digits.target 
```

```
data
```

```
target
```

##Manual Train Test Split

```
sample_number = len(data)
trainTestSplit = int(0.75 * sample_number)

model = SVC()
model.fit( data[:trainTestSplit], target[:trainTestSplit] )
```

##Prediction

```
prediction = model.predict(data[trainTestSplit:])
actual = target[trainTestSplit:]
```

##Model Evaluation

```
print('The confusion matrix is: ,')
print( confusion_matrix( actual, prediction ) )
print('The accuracy score is : ',accuracy_score(actual, prediction))
```

```
import seaborn as sns 
sns.heatmap( confusion_matrix(actual, prediction), annot = True, cmap = 'Blues' )
```

##Testing model through visualization

```
plt.imshow( digits.images[-68], cmap = plt.cm.gray_r, interpolation='nearest' )
print('The prediction for the figure is: ', model.predict(data[-68].reshape(1,-1)))
```
