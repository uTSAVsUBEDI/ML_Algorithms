##Polynomial Regression
#Importing Necessary Libraries 

```
import pandas as pd
from matplotlib import pyplot as plt 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
import math 
```

##Loading and reading dataset

```
df = pd.read_csv('/content/drive/My Drive/DataSets/Drugs_Performance.csv')
df.head(20)
```

##Data Frame to Array Conversion

```
X = df[["Age"]].values
y = df["Drugs_Perf"].values
print(X)
print(y)
```

##Data Visualization

```
plt.plot(X,y, c = 'orange')
plt.title('Data Visualization')
plt.xlabel('Age')
plt.ylabel('Drug Performance')
```

#Train Test Split

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

##Changing into polynomial Regression

```
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X_train)
poly.fit(X_poly, y_train)
```

##Training Model

```
model = LinearRegression()
model.fit(X_poly, y_train)
```

##Visualization

```
plt.plot(X,y, c = 'orange')
plt.title('Predicted Model')
plt.xlabel('Age')
plt.ylabel('Drugs Effects')
plt.plot(X, model.predict(poly.fit_transform(X)), c = 'k')
```

```
y_pred = model.predict(poly.fit_transform(X_test))
print(y_test)
print(';;;;;;;;;;')
print(y_pred)
```

##Model Evaluation

```
mse = mean_squared_error(X, y)
print("Mean Squared Error: ", math.sqrt(mse))
print("The R squard Error: ", r2_score(y_test,y_pred))
```
