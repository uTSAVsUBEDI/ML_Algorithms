#Support Vector Regression

#importing necesary libraries

```
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.svm import SVR
```

#Data Framing

```
df = pd.read_csv('/content/drive/My Drive/DataSets/Position_Salaries.csv')
df.head(10)
```

```
X = df.iloc[:,1:2].values.astype(float)
y = df.iloc[:,2:3].values.astype(float)
```

##Feature Scaling

```
sc = StandardScaler()
X = sc.fit_transform(X)
y = sc.fit_transform(y)
```

##Visualization of DataSet

```
plt.scatter(X, y, c= 'orange')
plt.title('Data Visualization')
plt.xlabel('Levels')
plt.ylabel('Salary')
```

##Building Model

```
model = SVR(kernel = 'rbf')
model.fit(X,y)
```

##Model Visualization

```
plt.scatter(X, y , c = 'orange')
plt.title('Model Building')
plt.xlabel('Levels')
plt.ylabel('salaries')
plt.plot(X, model.predict(X), c = 'blue')
```

##Model Prediction

```
y_pred = sc.inverse_transform((model.predict([[10]])))
print("The predicted Value is: ",y_pred)
```

##Model Accurency

```
import math
mse = mean_squared_error(X,y)
print("The Mean Squared Error is : ", math.sqrt(mse))
```
