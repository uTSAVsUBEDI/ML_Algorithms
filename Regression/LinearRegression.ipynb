# Linear Regression
##Importing necessary Libraries 



```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
import math
```

##DataSet

```
dataset = pd.read_csv("/content/drive/My Drive/DataSets/headbrain.csv")
dataset.head(6)
```

#Data Visualization

```
weight = dataset['Brain Weight(grams)']
size = dataset['Head Size(cm^3)']

plt.scatter(size, weight, c = 'orange')
plt.title('Size of brain with Weight')
plt.xlabel('Head size')
plt.ylabel('Brain Weight')
plt.show()
```

##Converting DataFrames into Numpy Arrays

```
X = np.array(size).reshape(-1,1)
y = np.array(weight).reshape(-1,1)
```

##Training Model


```
model = LinearRegression()
model.fit(X,y)
```

#Determining Errors

```
mse = mean_squared_error(X,y)
print("The MSE of the model is : ",math.sqrt(mse))
print("The R squared Error is : ", model.score(X,y))
```

#Fitting the Regression Line

```
plt.scatter(X,y, c = 'orange')
plt.plot(X, model.predict(X), c = 'b')
plt.title('Prediction Model')
plt.xlabel('Size of Brain')
plt.ylabel('Weight of Brain')
```

#Prediction

```
print("The predicted data for 2000 is: ", model.predict([[2000]]))
```
