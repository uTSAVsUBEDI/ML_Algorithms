##DBSCAN Clustering 

#Importing necessaery libraries 


```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
from sklearn.cluster import DBSCAN
from sklearn import datasets
```

#Loading dataset

```
x,y = datasets.make_moons(n_samples=2500, noise= 0.05)

x1 = x[:,0]
x2 = x[:,1]
```

#Visualizing DataSets


```
plt.scatter(x1,x2, s = 5)
plt.show()
```

##Finding optimal value of epsilon

```
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors( n_neighbors = 3 )
neigh = neigh.fit(x)
distances , indices = neigh.kneighbors(x)
```

```
distances = np.sort(distances, axis = 0)
distances = distances[:,1]
plt.plot(distances)
```

#Training with DBSCAN model

```
model = DBSCAN( eps = 0.05, min_samples = 5 )
model.fit(x,y)
y_pred = model.labels_.astype(int)
print(y_pred)
```

##Visualizing Formed model


```
colors = np.array(['#ff0000','#00ff00'])
print("Clusters:\n")
plt.scatter(x1,x2,s=5,color = colors [y_pred])
plt.show()
```
