##KMeans Clustering

##Importing necessary libraries

```
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import collections 
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
```

##Loading Dataset

```
df = pd.read_csv('/content/drive/My Drive/DataSets/Quotes.csv')
df
```

```
data = df['Quotes'].tolist()
data
```

##Defining a function Tokanizer 

```
def tokanizer( text ):
  token = word_tokenize(text)
  stemmer = PorterStemmer()

  normalized_words = [ stemmer.stem(i) for i in token if i not in stopwords.words('english') ]
  return normalized_words
```

##Defining a function cluster_Sentences

```
def cluster_sentences( n_sentences, k ):
  Tfidf_vectorizer = TfidfVectorizer( tokenizer=tokanizer, lowercase = True, stop_words= stopwords.words('english') )

  Tfidf_Matrix = Tfidf_vectorizer.fit_transform(n_sentences)

  kmeans = KMeans( n_clusters=  k )
  kmeans.fit(Tfidf_Matrix)


  clusters = collections.defaultdict(list)
 
  for i,labels in enumerate(kmeans.labels_): 
    clusters[labels].append(i)

  return dict(clusters)

```

##Testing our model

```
k = 7
clusters = cluster_sentences(data, k)


for cluster in clusters:
  print('\nCLUSTERS : ',cluster,'\n\t')

  for i,sentence in enumerate(clusters[cluster]):
    print(i+1,':','\t',data[sentence])
  
```
